```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
import random
from skimage import feature
```

```python
image = data.camera()
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        image[i,j] += random.randint(-20,20)
```

```python
plt.imshow(image, cmap = plt.cm.gray)
```

```python
mm = np.median(image)
print(mm)
binary = image > mm
plt.imshow(binary, cmap = plt.cm.gray)
```

```python
original_img = data.camera()
img_edges = feature.canny(original_img,sigma=2,)
plt.imshow(img_edges, cmap = plt.cm.gray)
```
